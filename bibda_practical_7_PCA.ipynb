Write a program to explain the different functions of the principal components



```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

```

```python
# Generate some random data
np.random.seed(0)
X = np.random.randn(100, 2)  # 100 samples, 2 features

```

```python
# Visualize the original data
plt.figure(figsize=(8, 4))
plt.scatter(X[:, 0], X[:, 1], alpha=0.6)
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

```

```python
# Perform PCA
pca = PCA(n_components=2)  # Retain both principal components
X_pca = pca.fit_transform(X)

```

```python
# Principal components
print("Principal Components:")
print(pca.components_)

```

```python
# Explained variance
print("\nExplained Variance:")
print(pca.explained_variance_)

```

```python
# Explained variance ratio
print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

```

```python
# Cumulative explained variance ratio
print("\nCumulative Explained Variance Ratio:")
print(np.cumsum(pca.explained_variance_ratio_))

```

```python
# Visualize principal components
plt.figure(figsize=(8, 4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.6)
plt.title('Data After PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

Explanation of different functions of principal components:

pca.components_: This attribute returns the principal axes in feature space, representing the directions of maximum variance in the data.

pca.explained_variance_: This attribute returns the amount of variance explained by each of the selected components.

pca.explained_variance_ratio_: This attribute returns the ratio of explained variance by each of the selected components. It represents the proportion of the dataset's variance that lies along each principal component.

np.cumsum(pca.explained_variance_ratio_): This calculates the cumulative explained variance ratio, showing how much of the total variance is explained as we use additional principal components.
